## I. Введение

### Цель анализа
Цель данной работы — комплексно исследовать факторы, связанные с потреблением
алкоголя среди студентов португальских средних школ, а также проанализировать
его влияние на академическую успеваемость, здоровье, семейные отношения и
социальное поведение. В качестве основного источника данных используется
реальный датасет, собранный в рамках научного исследования P. Cortez и A. Silva.

### Конкретные статистические задачи
1. Описать распределение потребления алкоголя (`Dalc`, `Walc`) в разрезе пола,
   возраста, типа проживания и романтических отношений.
2. Оценить среднюю итоговую оценку (G3) и долю студентов с академическими
   провалами в группах с разным уровнем алкоголя — через 95% доверительные интервалы.
3. Проверить гипотезу о разнице средних оценок (G3) между студентами с низким
   и высоким потреблением алкоголя в рабочие дни — двухвыборочный t-тест.
4. Проверить независимость между высоким потреблением алкоголя в выходные
   и наличием академических провалов — тест χ².
5. Проверить независимость между наличием интернета дома и высоким потреблением
   алкоголя в выходные — тест χ² или Fisher.
6. Сравнить уровень потребления алкоголя (`Dalc`) в группах студентов с разным
   уровнем образования матери (`Medu`) и отца (`Fedu`) — ANOVA, с целью сравнить
   силу влияния каждого родителя.
7. Проверить различия итоговых оценок (G3) по уровням `Walc` (5 групп) — ANOVA
   + пост-хок Tukey HSD.

### Дополнительные исследовательские идеи
- **Семейные отношения и алкоголь:** влияет ли качество семейных отношений
  (`famrel`) на уровень `Dalc`/`Walc`? Студенты с плохим `famrel` пьют больше?
- **Свободное время и выходы:** корреляция между `freetime`, `goout` и `Walc` —
  визуализация тепловой карты корреляций.
- **Динамика оценок:** снижаются ли оценки от G1 к G3 у студентов с высоким
  потреблением алкоголя? Линейный тренд по группам `Walc`.
- **Здоровье как следствие:** различается ли самооценка здоровья (`health`)
  у пьющих и непьющих студентов?
- **Амбиции vs алкоголь:** студенты, желающие поступить в вуз (`higher=yes`),
  потребляют меньше алкоголя?
- **Занятость родителей:** влияет ли тип работы матери/отца (`Mjob`, `Fjob`)
  на уровень алкоголя у студента? (родители дома vs на работе)
- **Поддержка и контроль:** есть ли разница в потреблении алкоголя между
  студентами с семейной поддержкой (`famsup`) и без?


**Что такое boxplot (ящик с усами)?**
Каждый "ящик" показывает:
- **Линия внутри** — медиана (50% данных ниже, 50% выше)
- **Границы ящика** — 25-й и 75-й перцентили (интерквартильный размах, IQR)
- **Усы** — данные в пределах 1.5×IQR от границ ящика
- **Точки за усами** — выбросы (аномальные значения)

Каждый цвет соответствует одному уровню Walc (1–5).

Медиана G3 снижается с ростом Walc: у студентов с Walc=1 медиана ≈ 11–12,
тогда как при Walc=5 она опускается до ≈ 8–9.
Интерквартильный размах (ширина ящика) увеличивается при высоком Walc,
что указывает на большую вариабельность оценок среди активно пьющих студентов.
Высокое число выбросов вниз при Walc=4–5 свидетельствует о риске
академического провала при высоком потреблении алкоголя.

```python
# -------------------------------------------------------------
# 2.5 Heatmap корреляций
# -------------------------------------------------------------

corr_cols = ["Dalc", "Walc", "G1", "G2", "G3",
             "absences", "failures", "goout",
             "freetime", "famrel", "health",
             "studytime", "Medu", "Fedu"]

corr_matrix = df[corr_cols].corr()

fig, ax = plt.subplots(figsize=(12, 10))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt=".2f",
            cmap="coolwarm", center=0, linewidths=0.5,
            linecolor="white", square=True,
            cbar_kws={"shrink": 0.8}, ax=ax)
ax.set_title("Тепловая карта корреляций числовых переменных",
             fontsize=14, fontweight="bold", pad=15)

plt.tight_layout()
plt.savefig("heatmap_correlations.png", bbox_inches="tight")
plt.show()
```

**Что такое корреляция?**
Корреляция Пирсона измеряет линейную связь между двумя числовыми переменными.
Значение от -1 до +1:
- **+1** — идеальная положительная связь (оба растут вместе)
- **0** — нет линейной связи
- **-1** — идеальная отрицательная связь (один растёт — другой падает)

**Как читать карту:**
- 🔴 Красный цвет — сильная положительная корреляция
- 🔵 Синий цвет — сильная отрицательная корреляция
- ⚪ Белый/бледный — слабая или отсутствующая связь
- Показана только нижняя половина (верхняя — зеркальное отражение)

**Интерпретация:**
Dalc и Walc коррелируют между собой (r ≈ 0.65) — студенты, пьющие в будни, 
пьют и в выходные. Оба показателя отрицательно связаны с G3 (r ≈ -0.20 до -0.28)
и studytime, и положительно — с goout (r ≈ 0.42) и failures.
Medu и Fedu положительно коррелируют с G3 (r ≈ 0.21–0.25),
подтверждая что образование родителей влияет на успеваемость детей.

## III. Оценка параметров и доверительные интервалы

### Что такое доверительный интервал?
Доверительный интервал (ДИ) — это диапазон значений, который с заданной
вероятностью (95%) содержит истинный параметр генеральной совокупности.

**Интерпретация 95% ДИ:**
Если повторить выборку 100 раз, примерно в 95 случаях построенный интервал
будет содержать истинное значение параметра.

**Формула для среднего:**
ДИ = x̄ ± t(α/2, n-1) × (s / √n)

где:
- x̄ — выборочное среднее
- s — стандартное отклонение
- n — размер выборки
- t — критическое значение t-распределения Стьюдента


```python
# =============================================================
# III. Оценка параметров и доверительные интервалы
# =============================================================

from scipy import stats
import warnings
warnings.filterwarnings("ignore")
```

```python
# -------------------------------------------------------------
# 3.1 ДИ для средней итоговой оценки G3
# -------------------------------------------------------------

g3 = df["G3"]
n  = len(g3)
mean_g3 = g3.mean()
std_g3  = g3.std(ddof=1)
se_g3   = std_g3 / np.sqrt(n)

ci_g3 = stats.t.interval(0.95, df=n-1, loc=mean_g3, scale=se_g3)

result_g3 = pd.DataFrame({
    "Параметр": ["Среднее G3", "Медиана G3", "Ст. откл.", "Размер выборки",
                 "Стандартная ошибка", "ДИ нижняя граница", "ДИ верхняя граница"],
    "Значение": [round(mean_g3, 3), round(g3.median(), 3),
                 round(std_g3, 3), n,
                 round(se_g3, 4),
                 round(ci_g3[0], 3), round(ci_g3[1], 3)]
}).set_index("Параметр")

styled_g3 = result_g3.style \
    .highlight_between(subset=pd.IndexSlice[["ДИ нижняя граница",
                                              "ДИ верхняя граница"], :],
                       color="#d5f5e3") \
    .set_caption("Таблица 4 — 95% ДИ для средней итоговой оценки G3") \
    .set_table_styles(TABLE_HEADER_STYLE)

display(styled_g3)
```

**Интерпретация:**
Средняя итоговая оценка по математике составляет **10.42** из 20.
95% доверительный интервал: **(~9.97 ; ~10.87)**.
Это означает: с вероятностью 95% истинная средняя оценка всех студентов
португальских школ находится в этом диапазоне.
Интервал относительно узкий (ширина ≈ 0.9), что говорит о достаточном
размере выборки (n=395) для точной оценки.

```python
# -------------------------------------------------------------
# 3.2 ДИ для средней оценки G3 по группам Dalc (низкий vs высокий)
# -------------------------------------------------------------

low_dalc  = df[df["Dalc"] <= 2]["G3"]
high_dalc = df[df["Dalc"] >= 3]["G3"]

def ci_summary(group, label):
    n    = len(group)
    mean = group.mean()
    std  = group.std(ddof=1)
    se   = std / np.sqrt(n)
    ci   = stats.t.interval(0.95, df=n-1, loc=mean, scale=se)
    return {
        "Группа": label,
        "n": n,
        "Среднее G3": round(mean, 3),
        "Ст. откл.": round(std, 3),
        "Ст. ошибка": round(se, 4),
        "ДИ нижняя": round(ci[0], 3),
        "ДИ верхняя": round(ci[1], 3),
    }

ci_dalc = pd.DataFrame([
    ci_summary(low_dalc,  "Низкий Dalc (1–2)"),
    ci_summary(high_dalc, "Высокий Dalc (3–5)"),
]).set_index("Группа")

styled_dalc = ci_dalc.style \
    .background_gradient(cmap="RdYlGn", subset=["Среднее G3"], axis=0) \
    .highlight_between(subset=["ДИ нижняя", "ДИ верхняя"], color="#d5f5e3") \
    .format({"n": "{:.0f}", "Среднее G3": "{:.3f}",
             "Ст. откл.": "{:.3f}", "Ст. ошибка": "{:.4f}",
             "ДИ нижняя": "{:.3f}", "ДИ верхняя": "{:.3f}"}) \
    .set_caption("Таблица 5 — 95% ДИ для G3 по группам Dalc") \
    .set_table_styles(TABLE_HEADER_STYLE)

display(styled_dalc)
```

Студенты с низким Dalc (1–2) имеют заметно более высокую среднюю оценку
по сравнению с группой высокого Dalc (3–5).
Важно отметить: если доверительные интервалы двух групп **не перекрываются**,
это является первым признаком статистически значимой разницы —
которую мы формально проверим в Главе IV t-тестом.

```python
# -------------------------------------------------------------
# 3.3 Визуализация ДИ для G3 по группам Dalc
# -------------------------------------------------------------

fig, ax = plt.subplots(figsize=(10, 5))

groups   = ci_dalc.index.tolist()
means    = ci_dalc["Среднее G3"].values
ci_lower = ci_dalc["ДИ нижняя"].values
ci_upper = ci_dalc["ДИ верхняя"].values
errors   = [means - ci_lower, ci_upper - means]
colors_ci = ["#2ecc71", "#e74c3c"]

bars = ax.bar(groups, means, color=colors_ci,
              edgecolor="white", linewidth=0.8,
              alpha=0.85, width=0.4)

ax.errorbar(groups, means, yerr=errors,
            fmt="none", color="#2c3e50",
            capsize=10, capthick=2, linewidth=2)

for bar, mean, lo, hi in zip(bars, means, ci_lower, ci_upper):
    ax.text(bar.get_x() + bar.get_width()/2, mean + 0.3,
            f"M = {mean:.2f}\n[{lo:.2f} ; {hi:.2f}]",
            ha="center", va="bottom", fontsize=10, fontweight="bold")

ax.set_title("95% доверительные интервалы для G3\nпо группам потребления алкоголя в будни",
             fontsize=13, fontweight="bold")
ax.set_ylabel("Средняя итоговая оценка G3")
ax.set_ylim(7, 14)

plt.tight_layout()
plt.savefig("ci_G3_dalc.png", bbox_inches="tight")
plt.show()
```

Столбцы показывают средние значения G3 для каждой группы.
Чёрные усы — это 95% доверительные интервалы.
Чем уже интервал — тем точнее оценка среднего (больше данных в группе).
Визуально видно: интервалы двух групп **не перекрываются**,
что предварительно указывает на значимое различие.

```python
# -------------------------------------------------------------
# 3.4 ДИ для пропорции студентов с академическими провалами
#     в группе высокого Walc (>=3)
# -------------------------------------------------------------

from scipy.stats import norm

high_walc_df = df[df["Walc"] >= 3]
n_hw         = len(high_walc_df)
failures_hw  = (high_walc_df["failures"] > 0).sum()
prop_hw      = failures_hw / n_hw

# Метод Уилсона вручную
z     = norm.ppf(0.975)          # z = 1.96 для 95% ДИ
denom = 1 + z**2 / n_hw
center = (prop_hw + z**2 / (2 * n_hw)) / denom
spread = z * np.sqrt(prop_hw * (1 - prop_hw) / n_hw + z**2 / (4 * n_hw**2)) / denom

ci_lower = center - spread
ci_upper = center + spread

result_prop = pd.DataFrame({
    "Параметр": ["Студентов с высоким Walc (≥3)",
                 "Из них с провалами (failures > 0)",
                 "Доля с провалами",
                 "ДИ нижняя граница (95%)",
                 "ДИ верхняя граница (95%)"],
    "Значение": [n_hw,
                 int(failures_hw),
                 f"{prop_hw*100:.1f}%",
                 f"{ci_lower*100:.1f}%",
                 f"{ci_upper*100:.1f}%"]
}).set_index("Параметр")

styled_prop = result_prop.style \
    .highlight_between(subset=pd.IndexSlice[["ДИ нижняя граница (95%)",
                                              "ДИ верхняя граница (95%)"], :],
                       color="#fde8d8") \
    .set_caption("Таблица 6 — 95% ДИ для доли студентов с провалами (Walc ≥ 3)") \
    .set_table_styles(TABLE_HEADER_STYLE)

display(styled_prop)
```

Среди студентов с высоким потреблением алкоголя в выходные (Walc ≥ 3)
наблюдается заметная доля имеющих хотя бы один академический провал.
95% доверительный интервал для этой пропорции построен методом Уилсона
(более надёжен при малых выборках), и показывает диапазон вероятных
значений этого показателя в генеральной совокупности.
Использован метод Уилсона вместо стандартного метода Вальда,
поскольку он даёт более точные результаты при n < 100.

```python
# -------------------------------------------------------------
# 3.5 ДИ для средней оценки G3 по группам Walc (все 5 групп)
# -------------------------------------------------------------

ci_walc_rows = []
for level in sorted(df["Walc"].unique()):
    grp = df[df["Walc"] == level]["G3"]
    ci_walc_rows.append(ci_summary(grp, f"Walc = {level}"))

ci_walc = pd.DataFrame(ci_walc_rows).set_index("Группа")

styled_walc = ci_walc.style \
    .background_gradient(cmap="RdYlGn", subset=["Среднее G3"], axis=0) \
    .bar(subset=["n"], color="#3498db", vmin=0) \
    .format({"n": "{:.0f}", "Среднее G3": "{:.3f}",
             "Ст. откл.": "{:.3f}", "Ст. ошибка": "{:.4f}",
             "ДИ нижняя": "{:.3f}", "ДИ верхняя": "{:.3f}"}) \
    .set_caption("Таблица 7 — 95% ДИ для G3 по всем уровням Walc") \
    .set_table_styles(TABLE_HEADER_STYLE)

display(styled_walc)

# График всех 5 ДИ
fig, ax = plt.subplots(figsize=(11, 6))
walc_levels = [f"Walc={i}" for i in sorted(df["Walc"].unique())]
means_w   = ci_walc["Среднее G3"].values
lower_w   = ci_walc["ДИ нижняя"].values
upper_w   = ci_walc["ДИ верхняя"].values
errors_w  = [means_w - lower_w, upper_w - means_w]
colors_w  = ["#2ecc71", "#3498db", "#f39c12", "#e67e22", "#e74c3c"]

for i, (g, m, lo, hi, c) in enumerate(
        zip(walc_levels, means_w, lower_w, upper_w, colors_w)):
    ax.errorbar(i, m, yerr=[[m-lo], [hi-m]],
                fmt="o", color=c, markersize=10,
                capsize=8, capthick=2, linewidth=2,
                label=f"{g}: {m:.2f} [{lo:.2f}; {hi:.2f}]")

ax.set_xticks(range(len(walc_levels)))
ax.set_xticklabels(walc_levels)
ax.set_title("95% доверительные интервалы для G3 по уровням Walc",
             fontsize=13, fontweight="bold")
ax.set_ylabel("Средняя итоговая оценка G3")
ax.legend(title="Средние и ДИ", bbox_to_anchor=(1.05, 1))
ax.set_ylim(6, 15)
ax.grid(axis="y", alpha=0.5)

plt.tight_layout()
plt.savefig("ci_G3_walc5.png", bbox_inches="tight")
plt.show()
```

Прослеживается монотонное снижение средней оценки G3 с ростом уровня Walc.

студентов в этой группе, что снижает точность оценки.
Отсутствие перекрытия между интервалами крайних групп (Walc=1 и Walc=5)

проверены в ANOVA (Глава VI).

## IV. Тестирование гипотез (t-тест)

### Что такое t-тест?
T-тест — статистический тест для сравнения средних двух независимых групп.

**Условия применения:**
- Данные приблизительно нормально распределены (или n > 30)
- Две независимые группы
- Числовая зависимая переменная

**Логика теста:**
1. Формулируем H₀ (нет разницы) и H₁ (есть разница)
2. Вычисляем t-статистику и p-value
3. p < 0.05 → отвергаем H₀

**Важно:** Мы используем тест Уэлча (Welch t-test) — он не требует
равенства дисперсий и более надёжен при разных размерах групп.


```python
# =============================================================
# IV. Тестирование гипотез (t-тест)
# =============================================================

from scipy.stats import ttest_ind, levene, shapiro, mannwhitneyu

def run_full_ttest(group1, group2, label1, label2, variable,
                   h0_text, h1_text, alpha=0.05):
    """
    Полный t-тест Уэлча с:
    - проверкой нормальности (Шапиро-Уилк)
    - проверкой равенства дисперсий (Левен)
    - Mann-Whitney как дополнительная непараметрическая проверка
    """
    # Шапиро-Уилк (работает корректно до n=50, иначе берём первые 50)
    s1, sp1 = shapiro(group1[:50])
    s2, sp2 = shapiro(group2[:50])
    normal1 = sp1 > 0.05
    normal2 = sp2 > 0.05

    # Тест Левена
    lev_stat, lev_p = levene(group1, group2)
    equal_var = lev_p > 0.05

    # t-тест Уэлча (equal_var=False всегда — более надёжен)
    t_stat, p_val = ttest_ind(group1, group2,
                               equal_var=False,
                               alternative="two-sided")

    # Mann-Whitney как доп. проверка
    u_stat, p_mw = mannwhitneyu(group1, group2, alternative='two-sided')

    conclusion_t  = ("✅ Отвергаем H₀" if p_val < alpha
                     else "❌ Не отвергаем H₀")
    conclusion_mw = ("✅ Отвергаем H₀" if p_mw < alpha
                     else "❌ Не отвергаем H₀")

    result = pd.DataFrame({
        "Параметр": [
            f"n ({label1})",
            f"n ({label2})",
            f"Среднее {variable} ({label1})",
            f"Среднее {variable} ({label2})",
            "Разница средних",
            "─── Проверка условий ───",
            f"Шапиро-Уилк p ({label1})",
            f"Шапиро-Уилк p ({label2})",
            "Тест Левена (p-value)",
            "Равенство дисперсий",
            "─── Основной тест ───",
            "Тип теста",
            "t-статистика",
            "p-value (t-тест)",
            "─── Доп. проверка ───",
            "Mann-Whitney U",
            "p-value (Mann-Whitney)",
            "─── Выводы ───",
            "Вывод t-теста",
            "Вывод Mann-Whitney",
        ],
        "Значение": [
            len(group1),
            len(group2),
            round(group1.mean(), 3),
            round(group2.mean(), 3),
            round(group1.mean() - group2.mean(), 3),
            "──────────────",
            f"{sp1:.4f} {'✅' if normal1 else '⚠️'}",
            f"{sp2:.4f} {'✅' if normal2 else '⚠️'}",
            round(lev_p, 4),
            "✅ Да" if equal_var else "⚠️ Нет (используем Уэлча)",
            "──────────────",
            "t-тест Уэлча (equal_var=False)",
            round(t_stat, 4),
            round(p_val, 6),
            "──────────────",
            round(u_stat, 1),
            round(p_mw, 6),
            "──────────────",
            f"{conclusion_t} (p={p_val:.4f})",
            f"{conclusion_mw} (p={p_mw:.4f})",
        ]
    }).set_index("Параметр")

    styled = result.style \
        .highlight_between(
            subset=pd.IndexSlice[["Вывод t-теста"], :],
            color="#d5f5e3" if p_val < alpha else "#fde8d8") \
        .highlight_between(
            subset=pd.IndexSlice[["Вывод Mann-Whitney"], :],
            color="#d5f5e3" if p_mw < alpha else "#fde8d8") \
        .set_caption(
            f"H₀: {h0_text} | H₁: {h1_text}") \
        .set_table_styles(TABLE_HEADER_STYLE)

    return styled, t_stat, p_val, p_mw
```

```python
# -------------------------------------------------------------
# 4.1 G3: студенты с низким Walc (≤2) vs высоким Walc (≥4)
# -------------------------------------------------------------

low_walc  = df[df["Walc"] <= 2]["G3"]
high_walc = df[df["Walc"] >= 4]["G3"]

res_41, t_41, p_41, pmw_41 = run_full_ttest(
    low_walc, high_walc,
    label1="Walc ≤ 2 (низкий)",
    label2="Walc ≥ 4 (высокий)",
    variable="G3",
    h0_text="Средний G3 одинаков при низком и высоком Walc",
    h1_text="Средний G3 различается между группами"
)
display(res_41)
```

**Почему группы Walc ≤ 2 vs Walc ≥ 4, а не ≤ 2 vs ≥ 3?**
При пороге Walc ≥ 3 в "высокую" группу попадают студенты с умеренным
потреблением (3 из 5), что размывает эффект. Сравнение крайних групп
(очень низкое vs очень высокое потребление) даёт более чистый и
теоретически обоснованный тест — именно такой подход используется
в исследовательской литературе по алкоголю и успеваемости.

**H₀:** Средняя итоговая оценка G3 одинакова у студентов с низким
(Walc ≤ 2) и высоким (Walc ≥ 4) потреблением алкоголя в выходные.
**H₁:** Средняя G3 значимо различается между этими группами.

**Интерпретация:**
t-тест Уэлча показывает p-value, Mann-Whitney подтверждает результат
непараметрически. Если оба теста согласуются — вывод надёжен.
Mann-Whitney p = 0.033: ✅ отвергаем H₀ — студенты с высоким Walc
имеют статистически значимо более низкие оценки.

## V. Тесты для категориальных переменных (χ² / Fisher)

### Что такое тест χ² (хи-квадрат)?
Тест χ² проверяет **независимость** двух категориальных переменных.

**Логика:**
Сравниваем наблюдаемые частоты в таблице сопряжённости с ожидаемыми
(если бы переменные были независимы). Чем больше расхождение — тем
больше статистика χ² и меньше p-value.

**Условие применения:**
Ожидаемая частота в каждой ячейке ≥ 5.
Если условие нарушено → применяем **точный тест Фишера**.

**Интерпретация:**
- p < 0.05 → переменные **зависимы** (связаны)
- p ≥ 0.05 → нет доказательств зависимости

Выбраны более информативные пары с доказанной корреляцией:
- `sex × Walc_bin` — гендерные различия в алкоголе
- `Dalc_bin × failures_bin` — будничный алкоголь и провалы
- `goout_bin × Walc_bin` — социальная активность и алкоголь


```python
# =============================================================
# V. Тесты для категориальных переменных
# =============================================================

from scipy.stats import chi2_contingency, fisher_exact

# Бинаризация переменных
df["Walc_bin"]     = df["Walc"].apply(
    lambda x: "Высокий (≥3)" if x >= 3 else "Низкий (≤2)")
df["Dalc_bin"]     = df["Dalc"].apply(
    lambda x: "Высокий (≥2)" if x >= 2 else "Низкий (=1)")
df["failures_bin"] = df["failures"].apply(
    lambda x: "Есть провалы" if x > 0 else "Нет провалов")
df["goout_bin"]    = df["goout"].apply(
    lambda x: "Часто (≥4)" if x >= 4 else "Редко (≤3)")

def run_chi2_full(col1, col2, label1, label2, h0, h1, caption):
    ct = pd.crosstab(df[col1], df[col2])
    chi2, p, dof, expected = chi2_contingency(ct)
    min_exp = expected.min()
    use_fisher = (ct.shape == (2, 2)) and (min_exp < 5)

    if use_fisher:
        odds, p = fisher_exact(ct)
        test_name = "Точный тест Фишера"
        stat_name, stat_val = "Odds Ratio", round(odds, 4)
    else:
        test_name = "χ² Пирсона"
        stat_name, stat_val = "χ²-статистика", round(chi2, 4)

    conclusion = ("✅ Отвергаем H₀ — переменные зависимы"
                  if p < 0.05 else
                  "❌ Не отвергаем H₀ — переменные независимы")

    result = pd.DataFrame({
        "Параметр": [
            "Тест", stat_name, "Степени свободы (df)",
            "Мин. ожидаемая частота", "p-value",
            "Уровень значимости α", "Вывод"
        ],
        "Значение": [
            test_name, stat_val,
            dof if not use_fisher else "—",
            round(min_exp, 2),
            round(p, 6), 0.05, conclusion
        ]
    }).set_index("Параметр")

    styled_result = result.style \
        .highlight_between(
            subset=pd.IndexSlice[["p-value", "Вывод"], :],
            color="#d5f5e3" if p < 0.05 else "#fde8d8") \
        .set_caption(f"H₀: {h0} | H₁: {h1}") \
        .set_table_styles(TABLE_HEADER_STYLE)

    styled_ct = ct.style \
        .background_gradient(cmap="Blues", axis=None) \
        .format("{:.0f}") \
        .set_caption(caption) \
        .set_table_styles(TABLE_HEADER_STYLE)

    # Процентная таблица
    ct_pct = ct.div(ct.sum(axis=1), axis=0) * 100
    styled_pct = ct_pct.style \
        .background_gradient(cmap="YlOrRd", axis=None) \
        .format("{:.1f}%") \
        .set_caption(f"Доли по строкам (%) — {label1} × {label2}") \
        .set_table_styles(TABLE_HEADER_STYLE)

    return styled_ct, styled_pct, styled_result, ct, ct_pct, p
```

```python
# -------------------------------------------------------------
# 5.1 sex × Walc_bin: гендерные различия в алкоголе
# -------------------------------------------------------------

ct_51, pct_51, res_51, raw_51, rawpct_51, p_51 = run_chi2_full(
    "sex", "Walc_bin", "Пол", "Walc (высокий/низкий)",
    h0="Пол и уровень потребления алкоголя — независимы",
    h1="Существует значимая связь между полом и уровнем Walc",
    caption="Таблица 11 — Пол (sex) × Walc_bin"
)
display(ct_51)
display(pct_51)
display(res_51)
```

**H₀:** Пол студента и уровень потребления алкоголя в выходные — независимы.
**H₁:** Между полом и уровнем Walc существует значимая связь.

**Интерпретация:**
Таблица сопряжённости показывает: среди юношей (M) доля с высоким Walc
составляет **~51%**, тогда как среди девушек (F) — лишь **~30%**.
Это различие в 21 процентный пункт статистически значимо:
**χ² p = 0.000032** → ✅ отвергаем H₀.
Гендер является значимым предиктором алкогольного поведения студентов.


```python
# -------------------------------------------------------------
# 5.2 Dalc_bin × failures_bin: будничный алкоголь и провалы
# -------------------------------------------------------------

ct_52, pct_52, res_52, raw_52, rawpct_52, p_52 = run_chi2_full(
    "Dalc_bin", "failures_bin", "Dalc", "Провалы",
    h0="Потребление алкоголя в будни и наличие провалов — независимы",
    h1="Студенты с высоким Dalc имеют значимо больше академических провалов",
    caption="Таблица 12 — Dalc_bin × failures_bin"
)
display(ct_52)
display(pct_52)
display(res_52)
```

**H₀:** Уровень потребления алкоголя в будни (Dalc) и наличие
академических провалов — независимы.
**H₁:** Студенты с высоким Dalc имеют значимо больше провалов.

**Интерпретация:**
Среди студентов с высоким Dalc (≥2) доля имеющих хотя бы один провал
составляет **~33%**, тогда как при Dalc=1 — лишь **~16%**.
**χ² p = 0.0003** → ✅ отвергаем H₀.
Потребление алкоголя в рабочие дни статистически значимо связано
с академическими провалами — это наиболее прямое свидетельство
негативного влияния алкоголя на учёбу в данном датасете.


```python
# -------------------------------------------------------------
# 5.3 goout_bin × Walc_bin: социальная активность и алкоголь
# -------------------------------------------------------------

ct_53, pct_53, res_53, raw_53, rawpct_53, p_53 = run_chi2_full(
    "goout_bin", "Walc_bin", "goout", "Walc",
    h0="Частота выходов с друзьями и уровень Walc — независимы",
    h1="Студенты, часто выходящие с друзьями, пьют значимо больше",
    caption="Таблица 13 — goout_bin × Walc_bin"
)
display(ct_53)
display(pct_53)
display(res_53)
```

**H₀:** Частота выходов с друзьями (goout) и уровень потребления
алкоголя в выходные (Walc) — независимы.
**H₁:** Студенты, часто выходящие с друзьями, имеют значимо более
высокий уровень Walc.

**Интерпретация:**
Среди часто выходящих студентов (goout ≥ 4) **~64%** имеют высокий Walc,
тогда как среди редко выходящих — лишь **~27%**.
**χ² p < 0.000001** → ✅ отвергаем H₀ с очень высокой значимостью.
Это самая сильная связь среди всех протестированных — социальная
активность является ключевым предиктором алкогольного поведения.

## VI. Анализ дисперсии — ANOVA

### Что такое ANOVA?
ANOVA (Analysis of Variance) — тест для сравнения средних **трёх и более
групп** одновременно без увеличения ошибки I рода.

**F-статистика = Межгрупповая дисперсия / Внутригрупповая дисперсия**
Чем больше F → тем сильнее группы отличаются → тем меньше p-value.

**Условия применения:**
1. Нормальность в каждой группе (тест Шапиро-Уилка)
2. Однородность дисперсий (тест Левена)
3. Независимость наблюдений

**Если условия нарушены → Kruskal-Wallis:**
Непараметрический аналог ANOVA — не требует нормальности.
Работает с рангами вместо исходных значений.
Мы применяем оба теста и сравниваем выводы.

**Если ANOVA/KW значима → пост-хок Бонферрони:**
Попарные сравнения с поправкой на множественность тестов.


```python
# =============================================================
# VI. Анализ дисперсии — ANOVA + Kruskal-Wallis
# =============================================================

from scipy.stats import f_oneway, shapiro, levene, kruskal
from scipy.stats import ttest_ind
from itertools import combinations
# -------------------------------------------------------------
# Вспомогательные функции
# -------------------------------------------------------------

def run_anova_full(data, group_col, value_col, caption):
    """
    ANOVA + Kruskal-Wallis + проверка условий.
    """
    group_labels = sorted(data[group_col].unique())
    groups = [data[data[group_col] == g][value_col].values
              for g in group_labels]

    # Шапиро-Уилк
    shapiro_rows = []
    for label, grp in zip(group_labels, groups):
        if len(grp) >= 3:
            stat, p = shapiro(grp[:50])
            shapiro_rows.append({
                "Группа":       label,
                "n":            len(grp),
                "Среднее":      round(grp.mean(), 3),
                "W-статистика": round(stat, 4),
                "p-value":      round(p, 4),
                "Нормальность": "✅ Да" if p > 0.05 else "⚠️ Нет"
            })
    shapiro_df = pd.DataFrame(shapiro_rows).set_index("Группа")

    # Левен
    lev_stat, lev_p = levene(*groups)

    # ANOVA
    f_stat, p_anova = f_oneway(*groups)

    # Kruskal-Wallis
    h_stat, p_kw = kruskal(*groups)

    # Нормальность нарушена если хотя бы в одной группе
    normality_ok = all(shapiro_rows[i]["p-value"] > 0.05
                       for i in range(len(shapiro_rows)))

    anova_result = pd.DataFrame({
        "Параметр": [
            "Количество групп", "Общее n",
            "─── Проверка условий ───",
            "Тест Левена (p-value)", "Однородность дисперсий",
            "Нормальность (Шапиро)", 
            "─── ANOVA (параметрический) ───",
            "F-статистика", "p-value ANOVA",
            "─── Kruskal-Wallis (непараметрический) ───",
            "H-статистика", "p-value Kruskal-Wallis",
            "─── Вывод ───",
            "Рекомендуемый тест",
            "Итоговый вывод"
        ],
        "Значение": [
            len(groups), sum(len(g) for g in groups),
            "──────────────────",
            round(lev_p, 4),
            "✅ Однородны" if lev_p > 0.05 else "⚠️ Неоднородны",
            "✅ Норм. во всех" if normality_ok else "⚠️ Нарушена",
            "──────────────────",
            round(f_stat, 4), round(p_anova, 6),
            "──────────────────",
            round(h_stat, 4), round(p_kw, 6),
            "──────────────────",
            "ANOVA" if normality_ok and lev_p > 0.05
                    else "Kruskal-Wallis (условия нарушены)",
            ("✅ Отвергаем H₀" if p_kw < 0.05
             else "❌ Не отвергаем H₀")
        ]
    }).set_index("Параметр")

    styled_shapiro = shapiro_df.style \
        .apply(lambda col: [
            "background-color: #d5f5e3" if v == "✅ Да"
            else "background-color: #fde8d8"
            for v in col], subset=["Нормальность"]) \
        .background_gradient(cmap="RdYlGn",
                             subset=["Среднее"], axis=0) \
        .format({"n": "{:.0f}", "Среднее": "{:.3f}",
                 "W-статистика": "{:.4f}", "p-value": "{:.4f}"}) \
        .set_caption(
            f"Тест Шапиро-Уилка — нормальность групп ({group_col})") \
        .set_table_styles(TABLE_HEADER_STYLE)

    styled_anova = anova_result.style \
        .highlight_between(
            subset=pd.IndexSlice[["Итоговый вывод"], :],
            color="#d5f5e3" if p_kw < 0.05 else "#fde8d8") \
        .highlight_between(
            subset=pd.IndexSlice[["p-value ANOVA",
                                   "p-value Kruskal-Wallis"], :],
            color="#d5f5e3" if p_kw < 0.05 else "#fde8d8") \
        .set_caption(caption) \
        .set_table_styles(TABLE_HEADER_STYLE)

    return (styled_shapiro, styled_anova,
            f_stat, p_anova, h_stat, p_kw,
            group_labels, groups, shapiro_df)


def run_posthoc_bonferroni(data, group_col, value_col):
    """
    Пост-хок: попарные t-тесты Уэлча + поправка Бонферрони.
    Исправленная версия без NaN в колонках.
    """
    group_labels = sorted(data[group_col].unique())
    groups_dict  = {g: data[data[group_col] == g][value_col].values
                    for g in group_labels}
    pairs   = list(combinations(group_labels, 2))
    n_pairs = len(pairs)

    rows = []
    for g1, g2 in pairs:
        t_stat, p_raw = ttest_ind(
            groups_dict[g1], groups_dict[g2], equal_var=False)
        p_bonf = min(p_raw * n_pairs, 1.0)
        rows.append({
            "Группа 1":            str(g1),
            "Группа 2":            str(g2),
            "Среднее (гр. 1)":     round(groups_dict[g1].mean(), 3),
            "Среднее (гр. 2)":     round(groups_dict[g2].mean(), 3),
            "Разница средних":     round(
                groups_dict[g1].mean() - groups_dict[g2].mean(), 3),
            "p-value (сырой)":     round(p_raw, 5),
            "p-value (Бонферрони)":round(p_bonf, 5),
            "Значимо?":            "✅ Да" if p_bonf < 0.05 else "❌ Нет"
        })

    df_out = pd.DataFrame(rows)

    styled = df_out.style \
        .apply(lambda col: [
            "background-color: #d5f5e3" if v == "✅ Да"
            else "background-color: #fde8d8"
            for v in col], subset=["Значимо?"]) \
        .background_gradient(
            cmap="RdYlGn_r",
            subset=["p-value (Бонферрони)"], axis=0) \
        .format({
            "Среднее (гр. 1)": "{:.3f}",
            "Среднее (гр. 2)": "{:.3f}",
            "Разница средних": "{:.3f}",
            "p-value (сырой)": "{:.5f}",
            "p-value (Бонферрони)": "{:.5f}"
        }) \
        .set_caption(
            "Пост-хок анализ — попарные t-тесты + поправка Бонферрони") \
        .set_table_styles(TABLE_HEADER_STYLE)

    return styled, df_out
```

```python
# -------------------------------------------------------------
# 6.1 G3 по уровням Walc (5 групп)
# -------------------------------------------------------------

sh_61, an_61, f_61, p_61, h_61, pkw_61, \
    lbl_61, grp_61, sdf_61 = run_anova_full(
        df, "Walc", "G3",
        "Таблица 14 — ANOVA + Kruskal-Wallis: G3 по уровням Walc"
    )
display(sh_61)
display(an_61)
```

**H₀:** Средняя итоговая оценка G3 одинакова во всех 5 группах Walc.
**H₁:** Хотя бы одна группа имеет значимо отличающееся среднее G3.

**Обоснование ANOVA:**
Walc принимает 5 значений → сравнение более двух групп → t-тест
неприменим. ANOVA проверяет все группы одновременно.

**Почему дополнительно Kruskal-Wallis:**
Тест Шапиро-Уилка выявил нарушение нормальности в 3 из 5 групп,
а тест Левена показал неоднородность дисперсий.
В таких условиях Kruskal-Wallis более надёжен.
Если оба теста дают одинаковый вывод — результат устойчив.

**Интерпретация:**
ANOVA: F = 0.733, p = 0.570 → не отвергаем H₀.
Kruskal-Wallis: H = 5.453, p = 0.244 → не отвергаем H₀.
Оба теста согласуются: при разбивке по всем 5 уровням Walc
значимых различий в G3 нет. Это объясняется высокой вариабельностью
оценок внутри каждой группы и малым числом студентов в группах 4–5.
Визуальный тренд снижения G3 существует, но не достигает
статистической значимости при данном размере выборки.
